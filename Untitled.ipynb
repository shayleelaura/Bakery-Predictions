# Popular Bakery Options

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import the data
bakery = pd.read_csv('Bakery.csv')
```

## Examine the Data

```python
#What do the columns and rows look like?
bakery.head()
```

```python
#What type of data types are in this data set
bakery.info()
```

## Data Exploration and Visualization

```python
#On average how many itmes are people buying
customers = bakery['TransactionNo'].nunique()
itemsSold = bakery['TransactionNo'].count()
avgSold = itemsSold/customers

print(customers)
print(itemsSold)
print(avgSold)
```

```python
#What are the different things sold in this bakery
bakery['Items'].value_counts()
bakery['Items'].unique()
```

```python
#What hours are most popular for the coffee shop

dateTime=pd.to_datetime(bakery['DateTime'])
bakery['Hour']=dateTime.dt.hour
```

```python
#Create bar plot of frequent hours

peakHours = bakery.groupby('Hour')['Items'].count()
peakHours.plot(kind = 'barh')
```

```python
#What items are purchased the most

itemFrequency = bakery['Items'].value_counts().sort_values(ascending=False)

itemFrequency.head(10).plot(kind = 'barh')
```
